```
import networkx as nx
import pylab as plt
from networkx.drawing.nx_agraph import graphviz_layout, to_agraph
from matplotlib.pyplot import figure
```

```
plt.rcParams['figure.figsize'] = [8.0, 5.0]
plt.rcParams['figure.dpi'] = 300
plt.style.use("fivethirtyeight")
```

```
G = nx.Graph()
G.add_edge('Ann Arbor', 'Detroit', weight=50)
G.add_edge('Ann Arbor', 'Toledo', weight=40)
G.add_edge('Detroit', 'Toledo', weight=60)
G.add_edge('Toledo', 'Cleveland', weight=120)
G.add_edge('Ann Arbor', 'Chicago', weight=260)
G.add_edge('Chicago', 'Fort Wayne', weight=148)
G.add_edge('Chicago', 'Indianapolis', weight=180)
G.add_edge('Fort Wayne', 'Indianapolis', weight=120)
G.add_edge('Indianapolis', 'Columbus', weight=180)
G.add_edge('Columbus', 'Toledo', weight=155)
G.add_edge('Columbus', 'Cleveland', weight=150)
G.add_edge('Columbus', 'Pittsburgh', weight=180)
G.add_edge('Pittsburgh', 'Philadelphia', weight=320)
```

```
pos = nx.spring_layout(G)
edge_labels = { (u,v): d['weight'] for u,v,d in G.edges(data=True) }

nx.draw_networkx_nodes(G,pos,node_size=100)
nx.draw_networkx_edges(G,pos)
nx.draw_networkx_labels(G,pos)
nx.draw_networkx_edge_labels(G,pos,edge_labels=edge_labels)
plt.axis('off')
plt.savefig('output.png',dpi=600)
plt.show()
```

```
print(nx.shortest_path(G, 'Philadelphia', 'Fort Wayne', weight='weight'))
```
